## Importing Libraries

```python
#importing libraries

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from datetime import datetime
import time
import csv
import pyautogui
import logging
```

## Writing Function for Scraping

```python
#initializing driver

driver = webdriver.Chrome()
driver.maximize_window()
driver.get('https://www.sofascore.com/tournament/football/england/premier-league/17#id:61627')

#creating lists
country = []
league = []
season = []
date = []
home_team = []
away_team = []
home_team_goals = []
away_team_goals = []
round_no = []
match_link = []

#moving page to the score table
webdriver.ActionChains(driver).move_to_element(driver.find_element(By.XPATH, '(//div[@class="Box Flex lcxsIJ jLRkRA"])[3]')).perform()

def scraping_scores():
    
    global country, league, season, date, home_team, away_team, home_team_goals, away_team_goals, round_no, match_link
    
#waiting for matches table appears and getting round back button
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//div[@class="Box kiSsvW"]')))
    round_back_button = driver.find_element(By.XPATH, '//button[@class = "Button iCnTrv"][1]')
    
#getting datas about the league
    country_name = driver.find_element(By.XPATH, '//span[@class = "Text eLjrJH"]').text
    league_name = driver.find_element(By.XPATH, '//h2[@class = "Text dnkMJl"]').text.split(' ')[0] + " " + driver.find_element(By.XPATH, '//h2[@class = "Text dnkMJl"]').text.split(' ')[1]
    season_year = driver.find_element(By.XPATH, '//div[@class = "Text kdfkXa"]').text
    
#getting current round number
    curr_round = driver.find_element(By.XPATH, '//div[@class = "Box kiSsvW"]//span[@class = "Text rJhVM"]').text.split(' ')[1]
    
#checking if all the matches during the round played or not and getting matches data
    while (int(curr_round) - 1) != 0:
#waiting for round number to appear
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//div[@class = "Box kiSsvW"]//span[@class = "Text rJhVM"]')))
        curr_round = driver.find_element(By.XPATH, '//div[@class = "Box kiSsvW"]//span[@class = "Text rJhVM"]').text.split(' ')[1]
        try:
#by checking position of the last game score in the table we can decide if all the matches played or not
            driver.find_element(By.XPATH, '//div[@class = "Box kiSsvW"]//a[@data-testid="event_cell"][10]//div[@data-testid="left_score"]')
            if int(curr_round) != 1:
#getting matches in the table
                matches = driver.find_elements(By.XPATH, '//div[@class = "Box kiSsvW"]//a[@data-testid="event_cell"]')
                for match in matches:
#getting datas from each matches and appeding them to the lists
                    country.append(country_name)
                    league.append(league_name)
                    season.append(season_year)
                    date.append(match.find_element(By.XPATH, './/bdi[@data-testid="event_time"]').text)
                    home_team.append(match.find_element(By.XPATH, './/div[@data-testid="left_team"]').text)
                    away_team.append(match.find_element(By.XPATH, './/div[@data-testid="right_team"]').text)
                    home_team_goals.append(match.find_element(By.XPATH, './/div[@data-testid="left_score"]').text)
                    away_team_goals.append(match.find_element(By.XPATH, './/div[@data-testid="right_score"]').text)
                    round_no.append(curr_round)
                    match_link.append(match.get_attribute('href'))
#adding pause to write the data to the lists
                time.sleep(3)
#clicking previous round button
                round_back_button.click()
            else:
#getting matches in the table
                matches = driver.find_elements(By.XPATH, '//div[@class = "Box kiSsvW"]//a[@data-testid="event_cell"]')
                for match in matches:
#getting datas from each matches and appeding them to the lists
                    country.append(country_name)
                    league.append(league_name)
                    season.append(season_year)
                    date.append(match.find_element(By.XPATH, './/bdi[@data-testid="event_time"]').text)
                    home_team.append(match.find_element(By.XPATH, './/div[@data-testid="left_team"]').text)
                    away_team.append(match.find_element(By.XPATH, './/div[@data-testid="right_team"]').text)
                    home_team_goals.append(match.find_element(By.XPATH, './/div[@data-testid="left_score"]').text)
                    away_team_goals.append(match.find_element(By.XPATH, './/div[@data-testid="right_score"]').text)
                    round_no.append(curr_round)
                    match_link.append(match.get_attribute('href'))
        except NoSuchElementException:
#clicking previous round button
            round_back_button.click()
            if int(curr_round) != 1:
#getting matches in the table
                matches = driver.find_elements(By.XPATH, '//div[@class = "Box kiSsvW"]//a[@data-testid="event_cell"]')
                for match in matches:
#getting datas from each matches and appeding them to the lists
                    country.append(country_name)
                    league.append(league_name)
                    season.append(season_year)
                    date.append(match.find_element(By.XPATH, './/bdi[@data-testid="event_time"]').text)
                    home_team.append(match.find_element(By.XPATH, './/div[@data-testid="left_team"]').text)
                    away_team.append(match.find_element(By.XPATH, './/div[@data-testid="right_team"]').text)
                    home_team_goals.append(match.find_element(By.XPATH, './/div[@data-testid="left_score"]').text)
                    away_team_goals.append(match.find_element(By.XPATH, './/div[@data-testid="right_score"]').text)
                    round_no.append(curr_round)
                    match_link.append(match.get_attribute('href'))
#adding pause to write the data to the lists
                time.sleep(3)
#clicking previous round button
                round_back_button.click()
            else:
#getting matches in the table
                matches = driver.find_elements(By.XPATH, '//div[@class = "Box kiSsvW"]//a[@data-testid="event_cell"]')
                for match in matches:
#getting datas from each matches and appeding them to the lists
                    country.append(country_name)
                    league.append(league_name)
                    season.append(season_year)
                    date.append(match.find_element(By.XPATH, './/bdi[@data-testid="event_time"]').text)
                    home_team.append(match.find_element(By.XPATH, './/div[@data-testid="left_team"]').text)
                    away_team.append(match.find_element(By.XPATH, './/div[@data-testid="right_team"]').text)
                    home_team_goals.append(match.find_element(By.XPATH, './/div[@data-testid="left_score"]').text)
                    away_team_goals.append(match.find_element(By.XPATH, './/div[@data-testid="right_score"]').text)
                    round_no.append(curr_round)
                    match_link.append(match.get_attribute('href'))
#initializing the function
scraping_scores()
```

## Showing the Data in DataFrame Format

```python
import pandas as pd
df = pd.DataFrame({'Country' : country, "League" : league, "Season" : season, "Round" : round_no, "Date" : date, "Home Team" : home_team, "Away Team" : away_team, 
                 "Home Team Goals" : home_team_goals, "Away Team Goals" : away_team_goals, "Match Page Link" : match_link})
```
